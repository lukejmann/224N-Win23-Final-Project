```python
import pandas as pd
import os

# open data/biden_data_large.csv
df = pd.read_csv('data/biden_data_large.csv')
# open data/random_data_small.csv
df2 = pd.read_csv('data/random_data_small.csv')

# only use the "Transcript Text" column of each
df = df[['Transcript Text']]
df2 = df2[['Transcript Text']]

# combine the two dataframes into one with label "pos" for biden and "neg" for random
df['label'] = 'pos'
df2['label'] = 'neg'
df = df.append(df2)

# preview the data with large columns not truncated
# shuffle the data
df = df.sample(frac=1).reset_index(drop=True)
pd.set_option('display.max_colwidth', None)
# show neg first
# df = df.sort_values(by=['label'])

# remove text all text prior to text that follows the format "UPPERCASE:  " where UPPERCASE is a single word of atleast 4 characters
df['Transcript Text'] = df['Transcript Text'].str.replace(r'^.*[A-Z]{4,}:  ', '', regex=True)
print(df)

# were going to to split save each row as a text file in either /train or /test
# we will use 80% of the data for training and 20% for testing

# file sturcture will be:
# data/
#   biden/
#     train/
#       pos/
#         0.txt
#         1.txt
#         ...
#       neg/
#         0.txt
#         1.txt
#         ...
#     test/
#       pos/
#         0.txt
#         1.txt
#         ...
#       neg/
#         0.txt
#         1.txt
#         ...


# create the train and test directories if they dont exist
if not os.path.exists('data/biden/train/pos'):
    os.makedirs('data/biden/train/pos')
if not os.path.exists('data/biden/train/neg'):
    os.makedirs('data/biden/train/neg')
if not os.path.exists('data/biden/test/pos'):
    os.makedirs('data/biden/test/pos')
if not os.path.exists('data/biden/test/neg'):
    os.makedirs('data/biden/test/neg')

# split the data into train and test
train = df[:int(len(df)*0.8)]
test = df[int(len(df)*0.8):]

# save the train data
for i in range(len(train)):
    row = train.iloc[i]
    with open('data/biden/train/' + row['label'] + '/' + str(i) + '.txt', 'w') as f:
        f.write(row['Transcript Text'])

# save the test data
for i in range(len(test)):
    row = test.iloc[i]
    with open('data/biden/test/' + row['label'] + '/' + str(i) + '.txt', 'w') as f:
        f.write(row['Transcript Text'])

        
```
